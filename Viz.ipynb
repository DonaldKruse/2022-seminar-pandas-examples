```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['figure.figsize'] = 16,8 # sets figure size 

df = pd.read_csv("fakedata.csv")
df.columns
```

```python
plt.figure()
ax = df.plot.scatter(x='t', y='A', color='r', label='A')
df.plot.scatter(x='t', y='B', color='g', label='B', ax=ax)
df.plot.scatter(x='t', y='C', color='b', label='C', ax=ax)

plt.show()
```

```python
plt.close()
df.drop('category', axis=1, inplace=True)
df.head()
```

```python
groupedTrials = df.groupby(['t'], as_index=False)[['A','B','C']]
means = groupedTrials.mean()
stds = groupedTrials.std()
```

```python
means
```

```python
stds
```

```python
plt.figure()
ax = means.plot(x='t', y='A', color='r', label='A')
means.plot(x='t', y='B', color='g', label='B', ax=ax)
means.plot(x='t', y='C', color='b', label='C', ax=ax)
plt.show()
```

```python
plt.figure()
ax = means.plot(x='t', y='A', yerr=stds['A'].to_numpy(), color='r', label='A')
means.plot(x='t', y='B', yerr=stds['B'].to_numpy(), color='g', label='B', ax=ax)
means.plot(x='t', y='C', yerr=stds['C'].to_numpy(), color='b', label='C', ax=ax)
plt.show()
```
